### Creating a List

A python list can be created by enclosing elements of similar or different data type in square brackets `[...]`, or with `range()` function.

````python
# Creating a list
week_days = ['Mon', 'Tue', 'Wed', 'Thur','Fri']
even_numbers = [2, 4, 6, 8, 10]
mixed_list = ['Mon', 1, 'Tue', 2, 'Wed', 3]

# Displaying elements of a list
print(week_days)
````

````python
# Creating a list with range()
nums = range(5)

for i in nums:
    print(i)
````

### Accessing the elements of the list

We can access the a given element of the list by providing the index of the element in a bracket. The index starts at 0 in Python.

````python
# Accessing the first elements of the list
print(week_days[0])
print(even_numbers[2])

# Getting the last element of the list
print(even_numbers[-1])
````

### Slicing a list

Given a list, we can slice it to get any parts or combination of its elements forming another list.

````python
# Get the elements from index 0 to 2. Index 2 is not included.
week_days = ['Mon', 'Tue', 'Wed', 'Thur','Fri']
print(week_days[0:2])

# Get elements from the last fourth elements to the first
# -1 starts at the last element 'Fri', -2 second last element `Thur'..... -4 to 'Tue'
print(week_days[-4:])

# Get all elements up to the fourth index
print(week_days[:4])

# Get all elements from the second to the last index
print(week_days[2:])

# You can use [:] to copy the entire list
print(week_days[:])
````

### Changing elements in a list

Python lists are mutable. We can delete or change the elements of the list.

````python
names = ['James', 'Jean', 'Sebastian', 'Prit']
print(names)

# Change 'Jean' to 'Nyandwi' and 'Sebastian' to 'Ras'
names[1:3] = ['Nyandwi', 'Ras']
print(names)

# Change 'Prit' to 'Sun'
names[-1] = 'Sun'
print(names)

# Change 'James' to 'Francois'
names[0] = 'Francois'
print(names)
````

````python
# Delete Nyandwi in names list using del
names = ['James', 'Jean', 'Sebastian', 'Prit']
del names[1]
print(names)
````

````python
# Using pop() to remove elements
names = ['James', 'Jean', 'Sebastian', 'Prit']
print(names.pop(2))
print(names)

# pop() without index specified will return the last element of the list
names = ['James', 'Jean', 'Sebastian', 'Prit']
print(names.pop())
print(names)
````

````python
# Using remove() to delete an element
names = ['James', 'Jean', 'Sebastian', 'Prit']
names.remove('James')
print(names)

# Using append() to add elements to the list
names = ['James', 'Jean', 'Sebastian', 'Prit']
names.append('Jac')
names.append('Jess')
print(names)
````

### Traversing a list

There are times that we may need to go over the list to read the elements of the list or perform iterative operations. We can use `for loop` to traverse through the list.

````python
# Given a list names, use for loop to display its elements
names = ['James', 'Jean', 'Sebastian', 'Prit']

for name in names:
    print(name)
````

````python
# Given a list nums, add 1 to the first element, 2 to the second, 3 to 3rd element, 4 to 4th element
# Example: nums = [1,2,3,6] will be nums_new = [2,4,6,10]

nums = [1, 2, 3, 6]
nums_new = []

for i in range(len(nums)):  # len(nums) gives the length of the list
    num = nums[i] + i + 1
    nums_new.append(num)

print(nums_new)
````

### Operations on list

#### Concatenating two lists

````python
# Concatenating two lists
a = [1,2,3]
b = [4,5,6]

c = a + b
print(c)

# Using * operator to repeat a list a number of times
print([None] * 5)
print([True] * 4)
print([1,2,4,5] * 2)
````

### Nested lists

#### Creating a list in other list

````python
# Creating a list in other list
nested_list = [1,2,3, ['a', 'b', 'c']]
print(nested_list)

# Get the ['a', 'b', 'c'] from the nested_list
print(nested_list[3])

# Indexing and slicing a nested list is quite similar to normal list
print(nested_list[1])
````

### List Methods

## 2.2 Dictionaries

Dictionaries are powerful Python builtin data structure that are used to store data of key and values.

- Creating a dictionary
- Accessing values and keys in dictionary

### Creating a dictionary

````python
# Creating a dictionary
countries_code = dict()
print(countries_code)

# Verify it's a dictionary
print(type(countries_code))

# Adding items to the empty dictionary
countries_code["United States"] = 1
print(countries_code)
````

````python
# Create a dictionary with {}
countries_code = {
        "United States": 1,
        "China": 86,
        "Rwanda":250,
        "Germany": 49,
        "India": 91,
}

print(countries_code)

# Add key and values to dictionary
countries_code['Australia'] = 61
print(countries_code)
````

### Accessing the values and keys in a dictionary

Just like how we get values in a list by using their indices, in dictionary, we can use a key to get its corresponding value.

````python
countries_code = {
    "United States": 1,
    "China": 86,
    "Rwanda": 250,
    "Germany": 49,
    "India": 91,
}

# Getting the code of Rwanda
print(countries_code["Rwanda"])

# Check if a key exists in a dictionary
print("India" in countries_code)
print("Singapore" in countries_code)
````

````python
# Getting the keys, values, and items of the dictionary
dict_keys = countries_code.keys()
dict_values = countries_code.values()
dict_items = countries_code.items()

print(f"Keys: {dict_keys}\nValues:{dict_values}\nItems:{dict_items}")

# Get the value of Australia
print(countries_code.get('Australia'))

# In case a provided key is absent...
print(countries_code.get('UK', 41))
````

## 3. Comparison and Logic operators

`Comparison` operators are used to compare values. It will either return true or false.

````python
# Greater than
print(100 > 1)

# Equal to
print(100 == 1)

# Less than
print(100 < 1)

# Greater or equal to
print(100 >= 1)

# Less or equal to
print(100 <= 1)

# String comparison
print('Intro to Python' == 'intro to python')
print('Intro to Python' == 'Intro to Python')
````

`Logic` operators are used to compare two expressions made by comparison operators.

* Logic `and` returns true only when both expressions are true, otherwise false.
* Logic `or` returns true when either any of both expressions is true. Only false if both expressions are false.
* Logic `not` as you can guess, it will return false when given expression is true, vice versa.

````python
# Logic and
print(100 == 100 and 100 == 100)
print(100 <= 10 and 100 == 100)

# Logic or
print(100 == 10 or 100 == 100)
print(100 == 10 or 100 == 10)

# Logic not
print(not 1 == 2)
print(not 1 == 1)
````

### 4.1 If, Elif, Else

Structure of If condition:

```
if condition:
  do something
else:
  do this
```

````python
if 100 < 2:
    print("As expected, nothing will be displayed")

if 100 > 2:
    print("This will be displayed")

if 100 < 2:
    print("As expected, nothing will be displayed")
else:
    print('Printed')
````

````python
# Assign ages and compare
john_age = 30
luck_age = 20

if john_age > luck_age:
    print("John is older than Luck")
else:
    print("John is younger than Luck")
````

````python
# Using multiple conditions with elif
john_age = 30
luck_age = 20
yan_age = 30

if john_age < luck_age:
    print("John is younger than Luck")
elif yan_age == luck_age:
    print("Yan's Age is same as Luck")
elif luck_age > john_age:
    print("Luck is older than John")
else:
    print("John's age is same as Yan")
````

````python
# Ternary operator examples
# Example 1: Return 'Even' if below num is 'Even' and 'Odd' if not
num = 45
result = 'Even' if num % 2 == 0 else 'Odd'
print(result)

# Example 2: Return True if a given element is in a list and False if not
nums = [1,2,3,4,5,6]
result = True if 3 in nums else False
print(result)
````

### 4.2 For Loop

For loop is used to iterate over list, string, tuples, or dictionary.

Structure of for loop:

```
for item in items:
  do something
```

````python
# For loop with list
even_nums = [2,4,6,8,10]

for num in even_nums:
    print(num)

# For loop with different list
week_days = ['Mon', 'Tue', 'Wed', 'Thur','Fri']

for day in week_days:
    print(day)
````

````python
# For loop with string (iterates over each character)
sentence = "It's been a long time learning Python. I am revisiting the basics!!"

for letter in sentence:
    print(letter)
````

````python
# For loop with split (iterate over words)
sentence = "It's been a long time learning Python. I am revisiting the basics!!"

for word in sentence.split():
    print(word)
````

````python
# For loop in dictionary
countries_code = { "United States": 1,
                 "India": 91,
                 "Germany": 49,
                 "China": 86,
                 "Rwanda":250
            }

for country in countries_code:
    print(country)

# For loop over dictionary values
for code in countries_code.values():
    print(code)
````

````python
# For loop with range
for number in range(10):
    print(number)

# For loop with range(start, end)
for number in range(10, 20):
    print(number)
````

````python
# List comprehension - traditional way
letters = []
for letter in 'MachineLearning':
    letters.append(letter)
print(letters)

# List comprehension - simplified
letters = [letter for letter in 'MachineLearning']
print(letters)
````

### 4.3 While loop

````python
a = 10
while a < 20:
    print('a is: {}'.format(a))
    a = a + 1
````

## 5. Functions

Functions are used to write codes or statements that can be used multiple times with different parameters.

One fundamental rule in programming is "DRY" or Do not Repeat Yourself. Function will help to not repeat yourself.

This is how you define a function in Python:

```
def function_name(parameters):

  """
  This is Doc String
  You can use it to notes about the functions
  """
  statements

  return results
```

* `function_name` is the name of the function. It must not be similar to any built in functions.
* `parameters` are the values that are passed to the function.
* `Doc String` is used to add notes about the function. It is not a must to use it but it is a `good practice`.
* `return` specify something or value that you want to return every time you call or run your function.

````python
# Function to add two numbers and return a sum
def add_nums(a, b):
    """
    Function to add two numbers given as inputs
    It will return a sum of these two numbers
    """
    sum_result = a + b
    return sum_result

print(add_nums(2, 4))
print(add_nums(4, 5))

# Displaying the doc string noted earlier
print(add_nums.__doc__)
````

````python
def activity(name_1, name_2):
    print("{} and {} are playing basketball!".format(name_1, name_2))

activity("Chris", "Francois")
activity("Kennedy", "Kleber")

# Note: Functions do not need to always have return. 
# When you only want to display the customized message (not involving expression), print() will be enough.
````

## 6. Lambda Functions

There are times that you want to create anonymous functions. These types of functions will only need to have one expression.

````python
# Sum of two numbers - traditional function
def add_nums_traditional(a, b):
    sum_result = a + b
    return sum_result

print(add_nums_traditional(1, 3))

# Sum of two numbers - using lambda
sum_of_two_nums = lambda c, d: c + d
print(sum_of_two_nums(4, 5))
````

### 7.1 Map function

Map gives you the ability to apply a function to an iterable structures such as list.

````python
def cubic(number):
    return number ** 3

num_list = [0,1,2,3,4]
print(list(map(cubic, num_list)))
````

### 7.2 Filter function

````python
def odd_check(number):
    return number % 2 != 0

# != is not equal to operation
num_list = [1,2,4,5,6,7,8,9,10,11]
print(list(filter(odd_check, num_list)))
````

### 8.1 List Comprehension

List comprehension makes it easy to make a new list from an existing list based on a given conditions.

````python
# Example: Get even numbers from a range
nums = range(1, 20)
even_nums = []

# Traditional way
for num in nums:
    if num % 2 == 0:
        even_nums.append(num)

print(even_nums)

# List comprehension way
even_nums = [num for num in nums if num % 2 == 0]
print(even_nums)
````

````python
# Example: Get days that start with 'T'
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
day_T = []

# Traditional way
for day in days:
    if day[0] == 'T':
        day_T.append(day)

print(day_T)

# List comprehension way
day_T = [day for day in days if day[0] == 'T']
print(day_T)
````

### 8.2 Enumerate Function

Enumerate function converts iterable objects into enumerate object. It basically returns a tuple that also contains a counter. That sounds hard, but with examples, you can see how powerful this function is...

````python
seasons = ['Spring', 'Summer', 'Fall', 'Winter']
print(list(enumerate(seasons)))

# The counter can start at a different value
print(list(enumerate(seasons, start=1)))

# Example with for loop
class_names = ['Rock', 'Paper', 'Scissor']
for index, class_name in enumerate(class_names, start=0):
    print(index, '-', class_name)
````

### 8.3 Zip Function

Zip is an incredible function that takes two iterators and returns a pair of corresponding elements as a tuple.

````python
name = ['Jessy', 'Joe', 'Jeannette']
role = ['ML Engineer', 'Web Developer', 'Data Engineer']

zipped_name_role = zip(name, role)
print(zipped_name_role)

# The zip object returns nothing. In order to show the zipped elements, we can use a list
print(list(zipped_name_role))
````
